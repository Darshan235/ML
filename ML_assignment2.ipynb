```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Heart_Disease_Prediction.csv')

print("First 5 rows of the dataset:")
display(df.head())

print("\nLast 5 rows of the dataset:")
display(df.tail())

print(f"\nNumber of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

print("\nColumn names:")
display(df.columns)
```

```python
print("\nDataset Information:")
df.info()

print("\nData Types of all columns:")
display(df.dtypes)

print("\nMissing values in each column:")
display(df.isnull().sum())

# Identify numerical and categorical features
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print(f"\nNumerical features: {numerical_cols}")
print(f"Categorical features: {categorical_cols}")
```

```python
# 1. Display statistical summary using describe()
print("\nStatistical Summary of the Dataset:")
display(df.describe())

# 2. Calculate mean, median, and standard deviation of age
age_mean = df['Age'].mean()
age_median = df['Age'].median()
age_std = df['Age'].std()

print(f"\nMean Age: {age_mean:.2f}")
print(f"Median Age: {age_median:.2f}")
print(f"Standard Deviation of Age: {age_std:.2f}")

# 3. Calculate minimum and maximum cholesterol values
cholesterol_min = df['Cholesterol'].min()
cholesterol_max = df['Cholesterol'].max()

print(f"\nMinimum Cholesterol: {cholesterol_min}")
print(f"Maximum Cholesterol: {cholesterol_max}")
```

```python
# Task 1: Count number of patients with and without heart disease
heart_disease_counts = df['Heart Disease'].value_counts()
print("\nNumber of patients with and without Heart Disease:")
display(heart_disease_counts)

# Task 2: Plot histogram for age
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=10, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Task 3: Plot histogram for cholesterol levels
plt.figure(figsize=(10, 6))
sns.histplot(df['Cholesterol'], kde=True, bins=15, color='lightcoral')
plt.title('Distribution of Cholesterol Levels')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Task 4: Interpret distributions (comments to guide interpretation)
print("\nInterpretation of Distributions:")
print("1. Heart Disease Counts: This shows the absolute frequency of 'Presence' and 'Absence' of heart disease in the dataset. A balanced or imbalanced dataset can be observed here.")
print("2. Age Distribution: The histogram for 'Age' visually represents the frequency of different age groups. Look for peaks (most common ages), spread (range of ages), and skewness (asymmetry of the distribution).")
print("3. Cholesterol Levels Distribution: The histogram for 'Cholesterol' similarly shows the frequency of various cholesterol levels. Observe the central tendency, spread, and any unusual patterns or outliers.")
```

```python
# Assuming 'numerical_cols' is already defined from previous steps
# If not, it can be defined as: numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

# Task 1: Compute correlation matrix for numerical columns
correlation_matrix = df[numerical_cols].corr()
print("\nCorrelation Matrix:")
display(correlation_matrix)

# Task 2: Identify highly correlated variables (e.g., correlation > 0.7 or < -0.7)
print("\nHighest correlated variable pairs (absolute correlation > 0.7):")
high_corr = correlation_matrix.abs().unstack()
high_corr = high_corr.sort_values(ascending=False)
high_corr = high_corr[high_corr > 0.7]
high_corr = high_corr[high_corr < 1.0] # Exclude self-correlation

# Filter to show unique pairs (A-B and B-A are the same)
high_corr_unique = high_corr.index.map(frozenset).unique()

for pair in high_corr_unique:
    col1, col2 = tuple(pair)
    if col1 in correlation_matrix.columns and col2 in correlation_matrix.columns:
        print(f"  - {col1} and {col2}: {correlation_matrix.loc[col1, col2]:.2f}")


# Task 3: Plot scatter plot between age and maximum heart rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Max HR', alpha=0.7)
plt.title('Scatter Plot of Age vs. Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Task 4: Interpret relationships
print("\nInterpretation of Relationships:")
print("1. Correlation Matrix: This matrix shows the Pearson correlation coefficient between all pairs of numerical variables. Values range from -1 to 1, where 1 indicates a perfect positive correlation, -1 a perfect negative correlation, and 0 no linear correlation.")
print("2. Highly Correlated Variables: Observing the pairs with high absolute correlation coefficients helps identify strong linear relationships. These could be useful for feature selection or understanding dependencies.")
print("3. Scatter Plot (Age vs. Max HR): This plot visually represents the relationship between a person's age and their maximum heart rate. Look for trends (e.g., does Max HR tend to decrease or increase with age?), clusters, or outliers. A negative correlation is generally expected, as maximum heart rate tends to decline with age.")
```
