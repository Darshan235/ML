```python
# 1. Convert score columns into NumPy arrays
math = df['math_score'].to_numpy()
science = df['science_score'].to_numpy()
english = df['english_score'].to_numpy()
overall = df['overall_score'].to_numpy()

# 2. Compute mean, median, standard deviation
stats = {
    'math': (np.mean(math), np.median(math), np.std(math)),
    'science': (np.mean(science), np.median(science), np.std(science)),
    'english': (np.mean(english), np.median(english), np.std(english)),
    'overall': (np.mean(overall), np.median(overall), np.std(overall))
}

# 3. Find minimum and maximum scores
min_max = {
    'math': (np.min(math), np.max(math)),
    'science': (np.min(science), np.max(science)),
    'english': (np.min(english), np.max(english)),
    'overall': (np.min(overall), np.max(overall))
}
print(f"maths: {math}\n science: {science}\n english: {english}\n overall: {overall}")
for subject, (mean, median, std) in stats.items():
    print(f"{subject.capitalize()}")
    print(f"  Mean   : {mean:.2f}")
    print(f"  Median : {median:.2f}")
    print(f"  Std Dev: {std:.2f}")
    print()
# Clean, readable print for min and max values
for subject, (min_val, max_val) in min_max.items():
    print(f"{subject.capitalize()}")
    print(f"  Min : {min_val}")
    print(f"  Max : {max_val}")
    print()
```

```python
# 1. Histogram of math scores
plt.figure()
plt.hist(df['math_score'])
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Histogram of Math Scores')
plt.show()

# 2. Bar chart of average scores
avg_scores = [
    df['math_score'].mean(),
    df['science_score'].mean(),
    df['english_score'].mean(),
    df['overall_score'].mean()
]

subjects = ['Math', 'Science', 'English', 'Overall']

plt.figure()
plt.bar(subjects, avg_scores)
plt.xlabel('Subjects')
plt.ylabel('Average Score')
plt.title('Average Scores by Subject')
plt.show()

# 3. Scatter plot between math and writing (english) scores
plt.figure()
plt.scatter(df['math_score'], df['english_score'])
plt.xlabel('Math Score')
plt.ylabel('English Score')
plt.title('Math vs English Scores')
plt.show()
```

```python
# Group scores by gender
gender_avg = df.groupby('gender')['overall_score'].mean()

# Group scores by parental education (correct column name)
parent_edu_avg = df.groupby('parent_education')['overall_score'].mean()

# Gender-wise bar chart
plt.figure()
plt.bar(gender_avg.index, gender_avg.values)
plt.xlabel('Gender')
plt.ylabel('Average Overall Score')
plt.title('Average Overall Score by Gender')
plt.show()

# Parental education-wise bar chart
plt.figure()
plt.bar(parent_edu_avg.index, parent_edu_avg.values)
plt.xlabel('Parental Education')
plt.ylabel('Average Overall Score')
plt.title('Average Overall Score by Parental Education')
plt.show()
```
